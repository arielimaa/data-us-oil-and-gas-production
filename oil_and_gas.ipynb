## Comparing with Crude Oil Production

```python
file = "data/U.S._crude_oil_production.csv"
oil_df = pd.read_csv(file, decimal=",", parse_dates=['Month'], skipinitialspace = True)
oil_df.head()
```

```python
oil_df.info()
```

## Yearly Oil production

```python
yearly_oil_df= oil_df.groupby(oil_df.Month.dt.year).sum()
yearly_oil_df
```

```python
yearly_oil_df.columns[1:].sort_values()
```

```python
yearly_oil_df.columns = yearly_oil_df.columns.str.strip()
```

```python
plot2 = yearly_oil_df.filter(items=['U.S. Crude Oil']).plot(kind="bar", figsize=(5,4))
plot2.set_xlabel("Year");
```

```python
filtered_yearly_oil_df= yearly_oil_df[np.logical_and(yearly_oil_df.index> 2008, yearly_oil_df.index<2018)]
```

### Check your code

## Merging Oil and Gas Production

## Merging Oil and Gas Production

```python
total_gas = filtered_yearly_gas_df[['U.S.']]
total_gas = total_gas.rename(columns={'U.S.': 'Gas'})
total_gas
```

```python
total_oil= filtered_yearly_oil_df[['U.S. Crude Oil']]
total_oil = total_oil.rename(columns={'U.S. Crude Oil': 'Crude Oil'})
total_oil
```

```python
merged_df = pd.concat([total_gas, total_oil], axis=1, join= 'inner')
merged_df
plot = merged_df.plot(kind="bar")
plot.set_xlabel("Year")
plot.legend(['Gas (Millions of Cubic feet)', 'Crude Oil (Thousands of barrels)']);
```
